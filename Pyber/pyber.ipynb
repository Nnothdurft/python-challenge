```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
cityData = pd.read_csv(city_data_to_load)
rideData = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
combinedData = pd.merge(rideData, cityData)
# Display the data table for preview
meanData  = combinedData.groupby("city").mean()
rideCount = combinedData.groupby("city").count()
meanData = meanData.drop(columns=["ride_id"])
plotData = pd.merge(meanData, combinedData, how="left", on="city")
plotData = plotData.drop(columns=["date", "fare_y", "ride_id", "driver_count_y"])
plotData = plotData.drop_duplicates("city")
plotData = plotData.join(rideCount["type"], rsuffix="count", on="city")
urban = plotData[plotData.type == "Urban"]
rural = plotData[plotData.type == "Rural"]
suburban = plotData[plotData.type == "Suburban"]
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Build the scatter plots for each city types
plt.scatter(urban["typecount"], urban["fare_x"], s=urban["driver_count_x"]*10,c="coral", edgecolors="black", alpha=.8)
plt.scatter(suburban["typecount"], suburban["fare_x"], s=suburban["driver_count_x"]*10,c="lightblue", edgecolors="black", alpha=.8)
plt.scatter(rural["typecount"], rural["fare_x"], s=rural["driver_count_x"]*10,c="gold", edgecolors="black", alpha=.8)
# Incorporate the other graph properties
plt.grid()
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
# Create a legend
plt.legend(["Urban", "Suburban", "Rural"], title="City Types")
plt.figtext(.92,.6, "Note:\nCircle size correlates with driver count per city")
# Incorporate a text label regarding circle size
# Save Figure
plt.savefig("rideShareData.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urbanFares = combinedData["fare"][combinedData.type == "Urban"].sum()
suburbanFares = combinedData["fare"][combinedData.type == "Suburban"].sum()
ruralFares = combinedData["fare"][combinedData.type == "Rural"].sum()
totalFares = urbanFares + suburbanFares + ruralFares
sizes = [urbanFares, ruralFares, suburbanFares]
# Build Pie Chart
plt.title("% of Total Fares by City Type")
plt.pie(sizes,explode=[.1,0,0], labels = ["Urban", "Rural", "Suburban"], autopct="%.1f%%", colors=["salmon", "gold", "lightblue"], startangle = 280, shadow=True)
# Save Figure
plt.savefig("faresByCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urbanRides = combinedData["fare"][combinedData.type == "Urban"].count()
suburbanRides = combinedData["fare"][combinedData.type == "Suburban"].count()
ruralRides = combinedData["fare"][combinedData.type == "Rural"].count()
totalRides = urbanRides + suburbanRides + ruralRides
# Build Pie Chart
sizes = [urbanRides, ruralRides, suburbanRides]
plt.title("% of Total Rides by City Type")
plt.pie(sizes,explode=[.1,0,0], labels = ["Urban", "Rural", "Suburban"], autopct="%.1f%%", colors=["salmon", "gold", "lightblue"], startangle = 260, shadow=True)
# Save Figure
plt.savefig("ridesByCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urbanDrivers = plotData["driver_count_x"][plotData.type == "Urban"].sum()
ruralDrivers = plotData["driver_count_x"][plotData.type == "Rural"].sum()
suburbanDrivers = plotData["driver_count_x"][plotData.type == "Suburban"].sum()
totalDrivers = urbanDrivers + ruralDrivers + suburbanDrivers
# Build Pie Charts
sizes = [urbanDrivers, ruralDrivers, suburbanDrivers]
plt.title("% of Total Drivers by City Type")
plt.pie(sizes,explode=[.1,0,0], labels = ["Urban", "Rural", "Suburban"], autopct="%.1f%%", colors=["salmon", "gold", "lightblue"], startangle = 230, shadow=True)
# Save Figure
plt.savefig("driversByCityType.png")
```

```python
# Show Figure
plt.show()
```
